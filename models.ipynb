# 🔍 K-Nearest Neighbors (KNN)

## What is KNN?

**K-Nearest Neighbors (KNN)** is a **non-parametric supervised learning algorithm** used for:

- Classification
- Regression

It is called a **lazy learning algorithm** because:

- It **does not build a model during training**
- It **stores the entire dataset**
- Computation happens **only when a prediction is required**

Instead of learning parameters, KNN predicts based on **distance relationships between samples**.

---

# 📐 Mathematical Representation

Assume we have a dataset:

$$
D = \{(x_1, y_1), (x_2, y_2), \dots , (x_n, y_n)\}
$$

Where:

- $x_i \in \mathbb{R}^d$ is a feature vector
- $y_i$ is the class label

Given a new data point:

$$
x_{new}
$$

KNN finds the **K closest samples** in the dataset.

The prediction is obtained by **majority voting**:

$$
\hat{y} = \operatorname{mode}(y_{(1)}, y_{(2)}, \dots , y_{(K)})
$$

Where:

- $y_{(k)}$ is the label of the **k-th nearest neighbor**

---

# 📏 Distance Metrics

Distance functions determine how similarity between points is measured.

---

## Euclidean Distance

The most commonly used distance metric.

$$
d(x,y) = \sqrt{\sum_{i=1}^{n}(x_i - y_i)^2}
$$

Where:

- $x_i$ represents the feature of point $x$
- $y_i$ represents the feature of point $y$
- $n$ is the number of dimensions

This measures the **straight-line distance in Euclidean space**.

---

## Manhattan Distance

Also known as **city-block distance**.

$$
d(x,y) = \sum_{i=1}^{n} |x_i - y_i|
$$

This measures the distance by **summing absolute differences across dimensions**.

---

## Minkowski Distance (General Form)

Both Euclidean and Manhattan distances are special cases of the **Minkowski distance**.

$$
d(x,y) = \left( \sum_{i=1}^{n} |x_i - y_i|^p \right)^{\frac{1}{p}}
$$

Where:

- $p = 1$ → Manhattan distance  
- $p = 2$ → Euclidean distance

---

# ⚙️ Choosing the Value of K

The hyperparameter **K** controls model complexity.

### Small K

- Highly flexible boundary
- Sensitive to noise
- High variance
- Risk of **overfitting**

### Large K

- Smooth decision boundary
- Lower variance
- Risk of **underfitting**

---

# 📊 Decision Boundary Intuition

KNN decision boundaries are **nonlinear** and adapt to the structure of the dataset.

Unlike parametric models, the boundary is determined directly by the **distribution of training samples**.

---

# 🚀 Support Vector Machines (SVM)

## What is SVM?

**Support Vector Machine (SVM)** is a powerful **supervised learning algorithm** used for:

- Classification
- Regression
- Outlier detection

SVM tries to find the **optimal separating hyperplane** that maximizes the margin between classes.

---

# 📏 Hyperplane Representation

For a linear classifier, the hyperplane is defined as:

$$
w^T x + b = 0
$$

Where:

- $w$ is the **weight vector**
- $x$ represents the **feature vector**
- $b$ is the **bias term**

This equation represents the **decision boundary**.

---

# 🧠 Maximum Margin Principle

SVM chooses the hyperplane that **maximizes the margin** between classes.

The margin is the distance between the hyperplane and the nearest training points.

These nearest points are called:

**Support Vectors**

The margin can be expressed as:

$$
\text{Margin} = \frac{2}{\|w\|}
$$

Maximizing the margin improves **generalization ability**.

---

# 📉 SVM Optimization Objective

The SVM optimization problem is:

$$
\min_{w,b} \frac{1}{2} ||w||^2
$$

Subject to:

$$
y_i(w^T x_i + b) \geq 1
$$

This ensures that all samples are **correctly classified with margin ≥ 1**.

---

# ⚙️ Hyperparameter C (Regularization)

The **C parameter** controls the trade-off between:

- Maximizing the margin
- Minimizing classification errors

The optimization problem becomes:

$$
\min_{w,b} \frac{1}{2}||w||^2 + C \sum_{i=1}^{n} \xi_i
$$

Where:

- $\xi_i$ are **slack variables** allowing misclassification
- $C$ penalizes errors

---

### Small C

- Allows misclassification
- Larger margin
- Simpler model
- Better generalization

---

### Large C

- Penalizes errors strongly
- Smaller margin
- Complex decision boundary
- Risk of **overfitting**

---

# 🌊 Gamma (γ) Parameter

Gamma appears in the **RBF kernel** and controls the influence of individual points.

$$
K(x_i,x_j) = \exp(-\gamma \|x_i - x_j\|^2)
$$

Where:

- $\gamma$ determines how far the influence of a training point reaches.

---

### Low Gamma

- Large influence radius
- Smooth boundary
- Risk of **underfitting**

---

### High Gamma

- Small influence radius
- Complex boundary
- Risk of **overfitting**

---

# 🎭 Kernel Trick

Many datasets are **not linearly separable**.

The **kernel trick** maps data into a higher-dimensional space where linear separation becomes possible.

Instead of explicitly computing the transformation:

$$
\phi(x)
$$

We compute inner products using:

$$
K(x_i,x_j) = \phi(x_i)^T \phi(x_j)
$$

---

# 🔧 Common Kernel Functions

## Linear Kernel

$$
K(x_i,x_j) = x_i^T x_j
$$

Best for:

- High dimensional data
- Text classification
- Large feature spaces

---

## Polynomial Kernel

$$
K(x_i,x_j) = (x_i^T x_j + c)^d
$$

Where:

- $d$ = polynomial degree
- $c$ = constant coefficient

This creates **curved decision boundaries**.

---

## RBF (Gaussian) Kernel

$$
K(x_i,x_j) = \exp(-\gamma \|x_i - x_j\|^2)
$$

Advantages:

- Handles complex nonlinear patterns
- Works well in many real-world datasets

---

## Sigmoid Kernel

$$
K(x_i,x_j) = \tanh(\alpha x_i^T x_j + c)
$$

This kernel behaves similarly to **neural network activation functions**.

---

# 📊 Comparison: KNN vs SVM

| Feature | KNN | SVM |
|------|------|------|
| Model Type | Instance-based | Optimization-based |
| Training Time | Very fast | Slower |
| Prediction Time | Slow | Fast |
| Handles Nonlinearity | Limited | Yes (with kernels) |
| Key Hyperparameters | K | C, Gamma, Kernel |

---

# 🧠 Key Takeaways

- **KNN** is simple and intuitive but computationally expensive during prediction.
- **SVM** finds an optimal separating boundary by solving an optimization problem.
- **Support vectors** define the decision boundary.
- **C** controls the trade-off between margin size and classification error.
- **Gamma** controls the flexibility of nonlinear decision boundaries.
- **Kernel functions** allow SVM to solve complex nonlinear problems.

```python
# =========================================================
# KNN vs SVM - Complete Intuition + Visualization Notebook
# Author: Abdelrhman Hassan
# =========================================================

# =============================
# 1. Imports
# =============================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_moons, make_blobs
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

sns.set_style("darkgrid")
np.random.seed(42)
```

Generate Dataset

```python
X, y = make_moons(n_samples=300, noise=0.2, random_state=42)

plt.figure(figsize=(7,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', edgecolors='k')
plt.title("Dataset Distribution")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

KNN Model

```python
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)

knn_acc = accuracy_score(y_test, y_pred_knn)

print("KNN Accuracy:", knn_acc)
```

KNN Evaluation

```python
cm = confusion_matrix(y_test, y_pred_knn)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues")

plt.title("KNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()

print(classification_report(y_test, y_pred_knn))
```

KNN Decision Boundary

```python
def plot_decision_boundary(model, X, y, title):

    h = 0.02

    x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
    y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1

    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, h),
        np.arange(y_min, y_max, h)
    )

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(7,6))

    plt.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")
    plt.scatter(X[:,0], X[:,1], c=y, edgecolors='k', cmap="coolwarm")

    plt.title(title)

    plt.show()


plot_decision_boundary(knn, X_train, y_train, "KNN Decision Boundary")
```

Support Vector Machine

```python
svm = SVC(kernel='rbf', C=1, gamma='scale')

svm.fit(X_train, y_train)

y_pred_svm = svm.predict(X_test)

svm_acc = accuracy_score(y_test, y_pred_svm)

print("SVM Accuracy:", svm_acc)
```

SVM Evaluation

```python
cm = confusion_matrix(y_test, y_pred_svm)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap="Reds")

plt.title("SVM Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()

print(classification_report(y_test, y_pred_svm))
```

SVM Decision Boundary

```python
plot_decision_boundary(svm, X_train, y_train, "SVM Decision Boundary")
```

Hyperparameter Tuning

```python
param_grid = {

    "C":[0.1,1,10,100],
    "gamma":[0.1,0.01,0.001],
    "kernel":["rbf"]

}

grid = GridSearchCV(SVC(), param_grid, cv=5)

grid.fit(X_train, y_train)

print("Best Params:", grid.best_params_)
```

Model Comparison

```python
models = ["KNN","SVM"]
scores = [knn_acc, svm_acc]

plt.figure(figsize=(6,5))

plt.bar(models, scores)

plt.title("Model Comparison")
plt.ylabel("Accuracy")

plt.ylim(0,1)

for i,v in enumerate(scores):
    plt.text(i, v+0.02, str(round(v,3)), ha="center")

plt.show()
```
